```python
%matplotlib notebook
import flashytlib.plot as fyplt
import flashytlib.calculator as fycal
import numpy as np

BaseDir = '/Users/ranchu/FLASHOR/Movies_n_Data/dele_1D_2S/'
ProblemDir = 'VX_G15_100ms_Fluid/VXG15100ms_CSSr_staticFluid_MI/'
Resol1    = '16X16E_1_30'
Resol2    = '16X20E_1_25'
Resol3    = '48X16E_1_30'
Resol4    = '48X20E_1_25'
Resol5    = '48X40E_1_07'
Resol6    = '48X40E_1_12'

filenamebase = '/deleptonizationwave_hdf5_chk_'
Labels = [Resol1,Resol2,Resol3,Resol4,Resol5,Resol6]
FileIndex    = [500,500,500,500,500,500]
LineType     = ['solid','dashdot', 'dashed','dotted','dashdot'] # linestyle='dashed'
Emin = 3.0
Emax = 300.0
ZoomE = [1.30,1.25,1.30,1.25,1.0748700581781865,1.12]

fnum = len(FileIndex)
AverageEnergy = fnum * ['?']
AverageFluxFactor = fnum * ['?']
Luminosity = fnum * ['?']
NumberDensity = fnum * ['?']
EnergyDensity = fnum * ['?']
J = fnum * ['?']
ECenter_J = fnum * ['?']
RadiusJ = fnum * ['?']
H1 = fnum * ['?']
ECenter_H = fnum * ['?']
RadiusH = fnum * ['?']

for i in range(fnum):
    fileindex = FileIndex[i]
    directory  = BaseDir + ProblemDir + Labels[i]  
    # ---
    fullfilename = directory + filenamebase + '{:04d}'.format(fileindex)
    print(fullfilename)
    
    [NumberDensity[i], EnergyDensity[i], AverageEnergy[i], J[i], ECenter_J[i], RadiusJ[i]] \
    = fycal.ReadMoment_Zeroth(fullfilename, Emin, Emax, ZoomE[i])
    [Luminosity[i], H1[i], ECenter_H[i], RadiusH[i]] \
    = fycal.ReadMoment_First(fullfilename,Emin, Emax, ZoomE[i])
    AverageFluxFactor[i] \
    = np.true_divide(Luminosity[i],EnergyDensity[i])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

LineType     = ['solid','dashed','dashdot','dotted']
fig, ax1 = plt.subplots(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
# --- Number Density ---
color = 'tab:red'
ax1.set_xlabel('Radius [cm]')
ax1.set_ylabel('Average Energy [MeV]',color=color)
lns = 2*fnum * ['?']
for i in range(fnum):
    leg = Labels[i]+ '_{:04d}'.format(FileIndex[i])
    if( i == 0):
        lns[i] = ax1.plot(RadiusJ[i],AverageEnergy[i],linestyle=LineType[i%4],color='black',label=leg)
    elif( i == fnum-1):
        lns[i] = ax1.plot(RadiusJ[i],AverageEnergy[i],linestyle=LineType[i%4],color='black',label=leg)
    else:
        lns[i] = ax1.plot(RadiusJ[i],AverageEnergy[i],linestyle=LineType[i%4],color=color,label=leg)

ax1.set_yscale('log')
ax1.set_xscale('linear') 
ax1.tick_params(axis='y', labelcolor=color)
# --- Average Flux Factor ---
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Average. Flux Factor', color=color)
for i in range(fnum):
    leg =  Labels[i]+ '_{:04d}'.format(FileIndex[i])
    if( i == 0):
        lns[i+fnum] = ax2.plot(RadiusH[i], AverageFluxFactor[i],linestyle=LineType[i%4], color='black',label=leg)
    elif( i == fnum-1):
        lns[i+fnum] = ax2.plot(RadiusH[i], AverageFluxFactor[i],linestyle=LineType[i%4], color='black',label=leg)
    else:
        lns[i+fnum] = ax2.plot(RadiusH[i], AverageFluxFactor[i],linestyle=LineType[i%4], color=color,label=leg)

ax2.set_yscale('linear')
ax2.tick_params(axis='y', labelcolor=color)
    
fig.tight_layout()  # otherwise the right y-label is slightly clipped
ax1.legend(loc='right')
#ax2.legend(loc='best')
ax1.set_title(ProblemDir)
plt.savefig(BaseDir+ProblemDir+'/DiffResol.png')

fig, ax1 = plt.subplots(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
# --- Luminosity ---
color = 'tab:red'
ax1.set_xlabel('Radius [cm]')
ax1.set_ylabel('Luminosity [MeV^3]',color=color)
lns = 2*fnum * ['?']
for i in range(fnum):
    leg = Labels[i]+'_{:04d}'.format(FileIndex[i])
    if(i == 0):
        lns[i] = ax1.plot(RadiusH[i],Luminosity[i],linestyle=LineType[i%4],color='black',label=leg)
    else:
        lns[i] = ax1.plot(RadiusH[i],Luminosity[i],linestyle=LineType[i%4],color=color,label=leg)

ax1.set_yscale('log')
ax1.set_xscale('log')
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend(loc='best')

plt.show()
```

```python
%matplotlib inline

base_i = 5
comp_i = 2

RelativeDiff_AverageEnergy= abs(AverageEnergy[base_i]-AverageEnergy[comp_i])/AverageEnergy[base_i]
RelativeDiff_AverageFluxFactor= abs(AverageFluxFactor[base_i]-AverageFluxFactor[comp_i])/abs(AverageFluxFactor[base_i])

import matplotlib.pyplot as plt

LineType     = ['solid','dashdot', 'dashed','dotted','dashdot']
#fig, ax1 = plt.subplots(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
fig, ax1 = plt.subplots(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
# --- Number Density ---
color = 'tab:red'
ax1.set_xlabel('Radius [cm]')
ax1.set_ylabel('Relative Diff. in Average Energy [MeV]',color=color)
lns = 2*fnum *['?']
leg = 'ABS('+ Labels[base_i]+'_{:04d}'.format(FileIndex[base_i])+' - '+ Labels[comp_i]+'_{:04d}'.format(FileIndex[comp_i])
leg = leg + ') / ' + Labels[base_i]+'_{:04d}'.format(FileIndex[base_i])
lns[0] = ax1.plot(RadiusJ[base_i],RelativeDiff_AverageEnergy,linestyle=LineType[0],color=color,label=leg)

ax1.set_yscale('log')
ax1.tick_params(axis='y', labelcolor=color)
# --- Average Flux Factor ---
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Relative Diff. in Average. Flux Factor', color=color)
lns[1] = ax2.plot(RadiusH[base_i], RelativeDiff_AverageFluxFactor,linestyle=LineType[1], color=color,label=leg)

ax2.set_yscale('log')    
ax2.tick_params(axis='y', labelcolor=color)
    
fig.tight_layout()  # otherwise the right y-label is slightly clipped
#ax1.legend(loc='upper left')
ax2.legend(loc='right')

plt.savefig(BaseDir+ProblemDir+'/DiffResol_RelaDiff.png')
plt.show()
```
