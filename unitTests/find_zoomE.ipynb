```python
%matplotlib inline 
                   # inline or notebook
# Test the Creation of Energy Grid
import flashytlib.calculator as fycal
import matplotlib.pyplot as plt
import numpy as np

xL = 3.
xR = 300.
N  = 40
nN = 2
Zoom = 1.07;
#Zoom = np.exp(np.log(xR-xL)/N) # suggested zoom
print('Zoom=', Zoom)
[center, width, NodesCoordinate] = fycal.CreateGeometricMesh( N, nN, xL, xR, Zoom)
# Gaussion [-0.28867513  0.28867513]

print('len(center)',len(center), ', min =',center[0],', max =',center[len(center)-1])
print('len(NodesCoordinate)',len(NodesCoordinate), 
      ', min =',NodesCoordinate[0])
print('NodesCoordinate value:', NodesCoordinate)

# =================   figure 1    =================
fig = plt.figure(num=None, figsize=(8, 3), dpi=80, facecolor='w', edgecolor='k')
ax2 = fig.add_subplot(111)
ax2.plot(center,np.linspace(1.,1.,num=N),'o',label='center')
ax2.set_yscale('linear')
ax2.set_xscale('log')
ax2.legend()

fig.suptitle('N = {:d},'.format(N)+' Zoom = {:3.3f}'.format(Zoom))
plt.show()

# =================    figure 2   =================
fig = plt.figure(num=None, figsize=(16, 3), dpi=80, facecolor='w', edgecolor='k')
ax1 = fig.add_subplot(121)
ax1.plot(width,'o',label='cell width')
ax1.set_yscale('log')
ax1.set_xscale('linear')
ax1.legend()
ax2 = fig.add_subplot(122)
ax2.plot(np.linspace(0.5,N-0.5,num=N),center,'o',label='center')
ax2.plot(np.linspace(0.25,N-0.25,num=N*nN),NodesCoordinate,'o',label='NodesCoordinate')
ax2.set_yscale('log')
ax2.set_xscale('linear')
ax2.legend()

fig.suptitle('N = {:d},'.format(N)+' Zoom = {:3.3f}'.format(Zoom))
plt.show()
```
